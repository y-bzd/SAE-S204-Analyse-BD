<h3 style="text-decoration : underline;">Informations Sur la Commodore C64/128/MAX<h3>

<p style="font-size : 20px; text-decoration: underline">La Commodore C64/128/MAX</p><p> est une console, plus precisement une machine mi-console mi-ordinateur ayant trois versions:</p> <p>- la Max Machine sortie en 1982 qui est une console de commodore derivée de la future commodore C64, une commodore C64 simplifiée et  bridée, et qui fut un echet immense.</p>
<p>- La C64 sortie en 1982 qui fut un immense succès, généralement considéré comme Première machine grand public hautement multimédia.</p>   <p>- 128 ou C128 sortie en 1989 qui est une version amélioré de la C64, avec plus de RAM, ... </p>
<p><strong>Commodore International</strong> est une entreprise américaine fondée dans les années 1950, qui a commencé à fabriquer des ordinateurs personnels à la fin des années 1970. Ses premières machines, comme le Commodore PET lancé en <strong>1977</strong>, formaient la première génération d’ordinateurs Commodore, surtout destinés à un usage professionnel et éducatif. La Commodore MAX Machine, sortie en <strong>1982</strong> au <strong>Japon</strong>, était une console basée sur la technologie du futur Commodore 64, mais elle a été rapidement arrêtée à cause de faibles ventes. Le Commodore 64, lancé la même année, est devenu un énorme succès populaire, vendu jusqu’en <strong>1994</strong>. En <strong>1985</strong>, le Commodore 128 est arrivé comme une version plus puissante, compatible avec le C64, marquant ainsi la deuxième et troisième génération des plateformes Commodore. Ces machines ont permis à Commodore de devenir un acteur majeur de l’informatique familiale dans les années 1980.</p>

```python
import getpass
from os import getenv
import pandas as pd
import oracledb
import warnings
import matplotlib.pyplot as plt
```

```python
# Nécessaire pour éviter les problèmes de session
class Connexion(object):
    def __init__(self, login, password):
        self.conn = oracledb.connect(
            user=login,
            password=password,
            host="oracle.iut-orsay.fr",
            port=1521,
            sid="etudom",
        )
        self.conn.autocommit = False

    def __enter__(self):
        self.conn.autocommit = False
        return self.conn

    def __exit__(self, *args):
        self.conn.close()
```

```python
def requete_vers_dataframe(connexion_data, requete, valeurs = None):
    with Connexion(login=connexion_data['login'], password=connexion_data['password']) as connexion:
        warnings.simplefilter(action='ignore', category=UserWarning)
        if valeurs is not None:
            df = pd.read_sql(requete, connexion, params=valeurs)
        else:
            df = pd.read_sql(requete, connexion)
        warnings.simplefilter("always") 
        return df

def requete_insert(connexion_data, requete, valeurs = None):
    with Connexion(login=connexion_data['login'], password=connexion_data['password']) as connexion:
        warnings.simplefilter(action='ignore', category=UserWarning)
        warnings.simplefilter("always")
        if valeurs is not None:
            df = pd.read_sql(requete, connexion, params=valeurs)
        else:
            df = pd.read_sql(requete, connexion)
        warnings.simplefilter("always")
```

<h3 style="text-decoration: underline;  text-align:center;">Répartition des jeux par catégories<h3>

```python
repartitionC = requete_vers_dataframe(conn,
f"""SELECT Count(idJeu) as Jeux, nomcategoriejeu FROM {SCHEMA}.PLATEFORME
NATURAL JOIN {SCHEMA}.DATESORTIE
NATURAL JOIN {SCHEMA}.JEU J
Inner Join {SCHEMA}.CategorieJeu C on C.idcategorieJeu=J.categorieJeu
where idplateforme={id_plateforme}
group by nomcategoriejeu
order by Count(idjeu) DESC
"""
)
#liste par comprehension contenant les nombres de jeux par catégories
nombreJeuParCategories=[repartitionC["JEUX"][i] for i in range (len(repartitionC))]
#liste par comprehension des catégories
categories=[repartitionC["NOMCATEGORIEJEU"][i] for i in range (len(repartitionC))]
s="     "
#affichage de la catégorie et de son nombre de jeu
for i in range(len(repartitionC)):
    print(s + repartitionC["NOMCATEGORIEJEU"][i] + " : " + str(repartitionC["JEUX"][i]))
    s+="  "
    
plt.figure(figsize=(12, 6))
#affiche le diagramme en bar verticale, avec les categories et nombre de jeux par categorie inversés
plt.barh(categories[::-1], nombreJeuParCategories[::-1], color='skyblue')
plt.xticks(rotation=90)
plt.title("Nombre de jeux par catégories")
plt.ylabel("Nombre de jeux")
plt.tight_layout()
plt.show()
```

<p style="color:steelblue;"><em>Sur la commodore C64/C128/MAX , on peut remarquer que la categorie de jeu la plus importantes est le main game , assez logique etant donné qu'on est dans l'ère de création de jeux vidéos qui marqueront l'histoire du jeu vidéo, etant donné le peu de jeu déja sortie et l'importance que consacre les gens aux jeux vidéos, il est normal de n'avoir quasiment aucun dlc, aucun remaster, etc...</em></p>

<h3 style="text-decoration: underline;  text-align:center;">Répartition des jeux par Theme<h3>

```python
repartitionT = requete_vers_dataframe(conn,
f"""SELECT Count(idJeu) as Jeux, nomtheme FROM {SCHEMA}.PLATEFORME
NATURAL JOIN {SCHEMA}.DATESORTIE
NATURAL JOIN {SCHEMA}.JEU
Natural Join {SCHEMA}.ThemeJeu
Natural Join {SCHEMA}.Theme
where idplateforme={id_plateforme}
group by nomtheme
order by Count(IdJeu) DESC
"""
)
#liste par comprehension contenant les nombres de jeux par theme
nombreJeuParTheme=[repartitionT["JEUX"][i] for i in range (len(repartitionT))]
#liste par comprehension des catégories
themes=[repartitionT["NOMTHEME"][i] for i in range (len(repartitionT))]
s=""
x=0
#affichage du theme et de son nombre de jeu associé
for i in range(len(repartitionT)):
    s+=  repartitionT["NOMTHEME"][i] + " : " + str(repartitionT["JEUX"][i]) + "   "
    x+=1
    if x%3==0:
        s+="\n"
    
print ("Theme et leur nombre de jeux :" + "\n\n" + s)

print("\n\n")   
plt.figure(figsize=(12, 6))
#affiche le diagramme en bar verticale, avec les themes et nombre de jeux par theme 
plt.bar(themes, nombreJeuParTheme, color='skyblue')
plt.xticks(rotation=90)
plt.title("Nombre de jeux par catégories")
plt.ylabel("Nombre de jeux")
plt.tight_layout()
plt.show()
```
